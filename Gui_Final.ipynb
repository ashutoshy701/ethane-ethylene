```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
from tkinter import *
from PIL import ImageTk,Image
from sklearn.preprocessing import MinMaxScaler

#####################################################################

def display_structure():
    labels = ['LCD:','PLD:', 'GSA:', 'AVF:', 'Density:']
    for i, label_text in enumerate(labels):
       label = Label(root, text=label_text,background="lightblue",font=("Arial MT Bold",10,"bold"),bg="white",relief="raised",bd=1)
       label.place(relx=0.6,rely=0.27+i*0.05)  # Align labels to the right (east)

       # Create an entry and display it beside the label
       entry = Entry(root,bd=2,bg="lightblue")
       entry.place(relx=0.7,rely=0.27+i*0.05) # Align entries to the left (west)
       entry_list.append(entry)
 


def display_chemicals():
    labels = ['TDU:', 'DU:','Metallic %:', 'OMR:','NOR:','H:', 'C:', 'N:', 'O:', 'F:', 'P:', 'S:', 'Cl:', 'Mn:', 'Co:', 'Ni:','Cu:', 'Zn:','Br:','Cd:', 'I:','Pb:']
    for i, label_text in enumerate(labels):
       label = Label(root, text=label_text,background="lightblue",font=("Arial MT Bold",10,"bold"),bg="white",relief="raised",bd=1)
       # Create an entry and display it beside the label
       entry = Entry(root,bd=2,bg="lightblue")
       if i < 13:
            label.place(relx=0.05, rely=0.26 + i * 0.05)
            entry.place(relx=0.14, rely=0.26 + i * 0.05)
            entry_list.append(entry)
       else:
            label.place(relx=0.3, rely=0.26 + (i - 13) * 0.05)
            entry.place(relx=0.35, rely=0.26 + (i - 13) * 0.05)
            entry_list.append(entry)



##########################################################################
root = Tk()

root.title("Selectivity and Ethane Uptake Prediction")
root.geometry("880x600") #size
root.minsize(400,300) #min size
root.maxsize(1000,1000) #max size



#root.iconbitmap("C:\\Users\\shrey\\OneDrive\\Documents\\project 2\\MOF\\mof_pic.ico")

#image = Image.open("C:\\Users\\shrey\\OneDrive\\Documents\\project 2\\MOF\\Case-Study-1.jpeg")
#resized_image=image.resize((100,60))
#photo =ImageTk.PhotoImage(image)
#Background_label=Label(root,image=photo)
#Background_label.place(x=0, y=0, relwidth=1,relheight=1)

header_font = ("Arial", 18, "bold")

# Create a header label
header_label = Label(root, text="Selectivity (C2H6/C2H4) and Ethane Uptake (mol/kg) Prediction", font=header_font,relief="groove", borderwidth=2)
header_label.pack(pady=20)
#label =Label(root,text ="Pridicting Selectivity and Uptake",font=("Playfair Display",20,"bold"),height=1)
#label.place(relx=0.5, rely=0.1, anchor=CENTER)


entry_list = []  # To keep references to the entry widgets

##################################################################################################################
def get_entry_values():
    global entry_values
    entry_values = []
    for entry in entry_list:
        try:
            value = float(entry.get())
            entry_values.append(value)
        except ValueError:
            # Handle the case where the user entered a non-numeric value
            entry_values.append(0.0)  # You can choose a default value or handle the error as needed
    print("Entry values:", entry_values)
####################################################################################################################
#label for showing predicted result
species_label1=Label(root,text ="",bg="White",font=("Arial",13,"bold"),width=10,borderwidth=2,relief="groove")
species_label1.place(relx=0.74,rely=0.76)
species_label2=Label(root,text ="",bg="White",font=("Arial",13,"bold"),width=10,borderwidth=2,relief="groove")
species_label2.place(relx=0.74,rely=0.86)
    
model = load_model("selectivity1.h5")
model2=load_model("uptake1.h5")


def pred_selec():
    b=np.asarray(entry_values)  
    b=np.reshape(b, (1, 27))
    p= model.predict(b) 
    p=p.round(2)
    species_label1.config(text=str(p[0,0]) )

def pred_Up():
    b=np.asarray(entry_values)  
    b=np.reshape(b, (1, 27)) 
    p2=model2.predict(b)
    p2=p2.round(2) 
    species_label2.config(text=str(p2[0, 0]))
####################################################################################################################
my_button1= Button(text="Chemical Descriptors",font=("Arial",12,"bold"),command=display_chemicals,relief="ridge",borderwidth=3)
my_button1.place(relx=0.15,rely=0.16)
my_button2= Button(text="Structural Descriptors",font=("Arial",12,"bold"),command=display_structure,relief="ridge", borderwidth=3)
my_button2.place(relx=0.62,rely=0.16)

my_button3= Button(text="Enter",font=("Arial",12,"bold"),command=get_entry_values,relief="ridge", borderwidth=3)
my_button3.place(relx=0.65,rely=0.55)

my_button4= Button(text="1. Selectivity (C2H6/C2H4):",font=("Arial",12,"bold"),command=pred_selec,relief="ridge", borderwidth=3)
my_button4.place(relx=0.55,rely=0.75)

my_button5= Button(text="2. Ethane Uptake (mol/kg):",font=("Arial",12,"bold"),command=pred_Up,relief="ridge", borderwidth=3)
my_button5.place(relx=0.55,rely=0.85)

print(entry_list)
root.mainloop()
```
